```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
```

```python
data = pd.read_csv('../0.csv',skiprows=1)
wspd = data.fillna(0)
wspd['date'] = pd.to_datetime(data[['Year', 'Month', 'Day', 'Hour', 'Minute']])
wspd = wspd[['date','wind speed at 100m (m/s)']]
wspd.set_index('date',inplace=True)
print(wspd.shape)
```

```python
res = STL(wspd).fit()
res.plot()
plt.show() 
```

```python
result = sm.tsa.seasonal_decompose(wspd,model='additive',period=288)
result.plot()
plt.show()
```
