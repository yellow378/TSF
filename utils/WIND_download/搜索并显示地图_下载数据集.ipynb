```python
import sys
import webbrowser
import time
import folium
import folium.plugins as plugins
import pandas as pd
import numpy as np
import math

from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
data=pd.read_csv('data/wtk_site_metadata.csv')
#选择特定洲郡的数据，也可以弄个查看全部的
State="California"
County="Siskiyou"
site=data.loc[(data['State']==State)&(data['County']==County),['latitude','longitude']]
site
```

```python
#从dataframe变成np.array数据
site1=np.asarray(site)
site1
```

```python
#打开加速器，加载地图
cal_map = folium.Map(location=site1[0])
for i in range(site1.shape[0]):
    folium.Marker(location=([site1[i][0],site1[i][1]]), icon=folium.Icon(color='blue', icon='cloud')).add_to(cal_map)

cal_map.add_child(folium.LatLngPopup()) 
draw = plugins.Draw()
draw.add_to(cal_map)
cal_map
```

```python
#根据得到的经纬度范围，筛选风机
polygon = Polygon([(-122.504018,41.492399), (-122.471389,41.423209), (-122.372128,41.450482), (-122.408192,41.517074)])
site2=np.empty(shape=[0,2])
for i in range(site1.shape[0]):
    point = Point(site1[i][1],site1[i][0])
    if(polygon.contains(point)):
        site2=np.append(site2,[site1[i]],axis=0)
site2
```

```python
#保存筛选得到的风机坐标
np.savetxt("data/zuobiao21.csv", site2, delimiter=',')
```

```python
cal_map = folium.Map(location=site2[0])
for i in range(site2.shape[0]):
    folium.Marker(location=([site2[i][0],site2[i][1]]), icon=folium.Icon(color='blue', icon='cloud')).add_to(cal_map)

cal_map.add_child(folium.LatLngPopup()) 
draw = plugins.Draw()
draw.add_to(cal_map)
cal_map
```
